```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
Y = []
X = []
h = 3
k = 4
n = 10000
x_range = 5.0
noise = np.random.normal(0,0.2,n)
for i in range(n):
    x = i*x_range/n
    X.append(x)
    Y.append(noise[i]+(x-h)**2+k)
```

```python
plt.plot(X, Y, 'ro')
plt.axis([0, xrange, 0, max(Y)])
plt.show()
```

```python
XY = list(zip(X,Y))
import random
random.shuffle(XY)
import pickle
with open('parabola-data.pkl','wb') as f:
    pickle.dump(XY, f)
```

```python
Y = []
X = []
n = 10000
x_range = 40.0
noise = np.random.normal(0,0.5,n)
for i in range(-n/2,n/2):
    x = i*x_range/n
    #print x
    Y.append(1 if abs(x) > 7 else 0)
    X.append(noise[i]+x)
```

```python
plt.plot(X, Y, 'ro')
plt.axis([-20, 20, 0, 2])
plt.show()
```

```python
XY = list(zip(X,Y))
import random
random.shuffle(XY)
import pickle
with open('step-data.pkl','wb') as f:
    pickle.dump(XY, f)
```
